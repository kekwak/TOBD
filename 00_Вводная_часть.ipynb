```python
# №4
def tl(n, m):
    return np.tile([[0, 1], [2, 3]], (m//2+1, n//2+1))[:m, :n]
    
tl(5, 4)
```

```python
# №7
desc = np.zeros((8,8))

desc[6,:].fill(1)
desc[7, 0::7] = 4
desc[7, 1::5] = 3
desc[7, 2::3] = 2
desc[7, 3:5] = (6, 5)

desc[::-1, ::-1]*-1 + desc
```
